## Grouping means:

Combining similar data into groups and performing calculations on them.

It is used when you want to:

find totals
averages
counts
maximum/minimum values

```python
import pandas as pd

data = {
    "Name": ["Yatin", "Rahul", "Aman", "Priya"],
    "Subject": ["Math", "Hindi", "Physic", "English"],
    "Marks": [90, 80, 70, 85]
}

df = pd.DataFrame(data)

group = df.groupby("Subject")["Marks"].mean()

print(group)
```
