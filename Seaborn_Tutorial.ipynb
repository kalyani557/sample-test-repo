# matplotlib → a Python library used for graphs and charts
# pyplot → a module inside matplotlib for plotting
## as plt → gives a short nickname plt

In Python, **Line** and Bar usually mean chart types used in data visualization with libraries like **Matplotlib or Pandas.**
Best for:

Temperature changes
Stock prices
Monthly sales trends

plt.plot() → makes a line chart ,

plt.bar() → makes a bar chart


```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 15, 30]

plt.plot(x, y)
plt.title("Line Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt

subjects = ["Math", "English", "Science"]
marks = [85, 70, 90]

plt.bar(subjects, marks)
plt.title("Bar Chart")
plt.show()
```

# A **histogram** is used to show how data is distributed (how values are spread).

```python
import matplotlib.pyplot as plt

marks = [40, 45, 50, 55, 60, 60, 65, 70, 75, 80, 85]

plt.hist(marks, bins=5)
plt.show()
```

# **plt.scatter()** is used to create a scatter plot.

A scatter plot shows the relationship between two sets of numbers using **dots.**

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 6]

plt.scatter(x, y)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

marks = [40, 45, 50, 55, 60, 60, 65, 70, 75]

sns.histplot(marks)

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 6]

sns.scatterplot(x=x, y=y)

plt.show()
```

```python
import numpy as np

data = np.array([[1,2],[3,4]])

sns.heatmap(data, annot=True)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Data
subjects = ["Math", "English", "Science", "History"]
marks = [85, 70, 90, 75]

# Theme + Grid
sns.set_style("darkgrid")

# Font style and scaling
sns.set_context("talk", font_scale=1)

# Figure size
plt.figure(figsize=(8, 5))

# Bar plot with palette colors
sns.barplot(
    x=subjects,
    y=marks,
    palette="viridis"
)

# Titles and labels
plt.title("Student Marks", fontsize=16)
plt.xlabel("Subjects", fontsize=12)
plt.ylabel("Marks", fontsize=12)

# Show graph
plt.show()
```

A box plot (or box-and-whisker plot) is a standardized graphical method for displaying the distribution of numerical data based on a five-number summary:

```python
import seaborn as sns
import matplotlib.pyplot as plt

marks = [40, 45, 50, 55, 60, 60, 65, 70, 75, 150]

sns.boxplot(x=marks)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file
df = pd.read_csv("/content/titanic.csv")

dead_people = df[df["survived"] == 0]

# Age section
age_30 = dead_people[(dead_people["age"] >= 30)
 & (dead_people["age"] < 40)]
age_40 = dead_people[(dead_people["age"] >= 40)
 & (dead_people["age"] < 50)]
age_50 = dead_people[(dead_people["age"] >= 50)
 & (dead_people["age"] < 60)]
counts = [
    len(age_30),
    len(age_40),
    len(age_50)
]

# Parts of pie charts
labels = [
    "30-39 Age",
    "40-49 Age",
    "50-59 Age"
]

# Size of pie chart
plt.figure(figsize=(7,7))

plt.pie(
    counts,
    labels=labels,
    autopct='%1.1f%%'
)

# Title
plt.title("Titanic Death Distribution by Age Group")
plt.show()
```
